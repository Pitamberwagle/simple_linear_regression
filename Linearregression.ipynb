```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv("Stipend_Data.csv")
X=dataset.iloc[:,0:1].values
Y=dataset.iloc[:,1].values
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=1/3,random_state=0)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
# Get the coefficients of the linear regression equation
slope = regressor.coef_[0]
intercept = regressor.intercept_
# Display the regression equation
equation = f'Regression Equation: Y = {slope:.2f} * X + {intercept:.2f}'
print(equation)
Y_predict=regressor.predict(X_test)
plt.scatter(X_train,Y_train,color="red")
plt.plot(X_train,regressor.predict(X_train),color="Blue")
plt.title("Training Data Analysis")
plt.xlabel("Years of experience")
plt.ylabel("Stipend")
plt.show()
plt.scatter(X_test,Y_test,color="Red")
plt.plot(X_train,regressor.predict(X_train),color="Blue")
plt.text(1,115000,equation)
plt.title("Test Data Analysis")
plt.xlabel("Years of experience")
plt.ylabel("Stipend")
plt.show()
from sklearn import metrics
print("Absolute error:", metrics.mean_absolute_error(Y_test,Y_predict))
print("Mean squared error:", metrics.mean_squared_error(Y_test,Y_predict))
print("RMS error:", np.sqrt(metrics.mean_squared_error(Y_test,Y_predict)))
```
